```python
#importing review text and label files to dataframe

import pandas as pd

df_test_text = pd.read_table('imdb_test_text.txt', delim_whitespace=False, names=('A'))
df_test_labels = pd.read_table('imdb_test_labels.txt', delim_whitespace=False, names=('B'))
```

```python
#merging them on index
df_test=df_test_text.join(df_test_labels, how='outer')
```

```python
#changing column name
df_test=df_test.rename(columns={'A':'review_text',
                          'B':'review_label'})
df_test.index.name='S.No.'
df_test
```

```python
df_train_text = pd.read_table('imdb_train_text.txt', delim_whitespace=False, names=('A'))
df_train_labels = pd.read_table('imdb_train_labels.txt', delim_whitespace=False, names=('B'))
```

```python
#merging them on index
df_train=df_train_text.join(df_train_labels, how='outer')
```

```python
#changing column name
df_train=df_train.rename(columns={'A':'review_text',
                          'B':'review_label'})
df_train.index.name='S.No.'
df_train
```

```python
frames=[df_train, df_test]
df_total=pd.concat(frames)
```

```python
df_total
```

```python
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics, svm
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import decomposition, ensemble

import pandas, numpy, string
```

```python
# create a count vectorizer object 
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(df_total['review_text'])

# transform the training and validation data using count vectorizer object
xtrain_count =  count_vect.transform(df_train)
xtest_count =  count_vect.transform(df_test)
```

```python
# word level tf-idf
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
tfidf_vect.fit(df_total['review_text'])
xtrain_tfidf =  tfidf_vect.transform(df_train)
xvalid_tfidf =  tfidf_vect.transform(df_test)

# ngram level tf-idf 
tfidf_vect_ngram = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', ngram_range=(2,3), max_features=5000)
tfidf_vect_ngram.fit(df_total['review_text'])
xtrain_tfidf_ngram =  tfidf_vect_ngram.transform(df_train)
xvalid_tfidf_ngram =  tfidf_vect_ngram.transform(df_test)

# characters level tf-idf
tfidf_vect_ngram_chars = TfidfVectorizer(analyzer='char', token_pattern=r'\w{1,}', ngram_range=(2,3), max_features=5000)
tfidf_vect_ngram_chars.fit(df_total['review_text'])
xtrain_tfidf_ngram_chars =  tfidf_vect_ngram_chars.transform(df_train) 
xvalid_tfidf_ngram_chars =  tfidf_vect_ngram_chars.transform(df_test) 
```
